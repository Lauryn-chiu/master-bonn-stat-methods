````python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import math 
from mpl_toolkits import mplot3d
%matplotlib widget 
from mpl_toolkits.mplot3d import Axes3D 
import matplotlib.ticker as mticker
````

````python
def sample_sphere(n):
    """
    Computes two random and uniform distribution for 
    the coordinates theta and phi. 
    
    Returns:
    """
    cos_theta = np.random.uniform(-1,1,n)
    # theta = np.arccos(cos_theta)
    theta= np.random.uniform(0,np.pi,n)
    phi = np.random.uniform(0,2*np.pi,n)

    return theta, phi, cos_theta
````

````python
sphere = sample_sphere(10000)

# sphere[0] -> array of theta coordinates
# sphere[1] -> array of phi coordinates
````

````python
# Just a function for the ticks in the angles
def format_func(value, tick_number):
    N = int(np.round(value / np.pi))
    if N == 0:
        return "0"
    elif N == 1:
        return r"$\pi$"
    elif N == -1:
        return r"-$\pi$"
    elif N == 2:
        return r"$2\pi$"
    elif N == -2:
        return r"-$2\pi$"
    else:
        return r"${0}\pi$".format(N)
````

````python
fig, axs = plt.subplots(ncols=3, figsize=(15, 5))

axs[0].hist(sphere[0], bins=100, color='c', alpha=0.7)
axs[0].set_title(r'$\theta$')
axs[0].set_xlabel('Position')
axs[0].set_ylabel('Frecuency')
axs[0].xaxis.set_major_locator(mticker.MultipleLocator(base=np.pi))
axs[0].xaxis.set_major_formatter(mticker.FuncFormatter(format_func))

axs[1].hist(sphere[2], bins=100, color='r', alpha=0.7)
axs[1].set_title(r'$\cos{\theta}$')
axs[1].set_xlabel('Position')
axs[1].set_ylabel('Frecuency')

axs[2].hist(sphere[1], bins=100, color='purple', alpha=0.7)
axs[2].set_title(r'$\phi$')
axs[2].set_xlabel('Position')
axs[2].set_ylabel('Frecuency')
axs[2].xaxis.set_major_locator(mticker.MultipleLocator(base=np.pi)) 
axs[2].xaxis.set_major_formatter(mticker.FuncFormatter(format_func))

# Ajustar espaciado entre los subplots
plt.tight_layout()

# Mostrar los gráficos
plt.show()
````

The following is just a coordinates change not a projection!

````python
def cartesian_coords(phi, theta, r = 1):
    x= r*np.sin(theta)*np.cos(phi)
    y= r*np.sin(theta)*np.sin(phi)
    z = r*np.cos(theta)
    return x, y, z
````

````python
cartesian_sphere = cartesian_coords(sphere[1],sphere[0])
````
