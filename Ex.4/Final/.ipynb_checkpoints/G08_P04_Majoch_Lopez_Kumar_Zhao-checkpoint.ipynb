## Problem 1: Finding the best-fit template spectrum amplitude

Let us assume that in restframe, all galaxies have a spectrum as the one given in `tspec[0]` but they vary in their total luminosity. Observationally we measure galaxy magnitudes, i.e. the logarithm of observed galaxy fluxes, in the five photometric bands given above. 

In this first problem, we will assume that we do not have a model for the measurement errors on galaxy magnitudes. As a first guess, we can therefore assume that the magnitudes (i.e. the log of the flux) are Gaussian-distributed with equal errors in each photometric band. Allowing for a variable galaxy luminosity in this model amounts to introducing a constant amplitude $A$ that consistently scales the model log-fluxes in all photometric bands.

(i) Write down an analytic expression for the likelihood that needs to be minimzed to find the best-fit amplitude $A$ for a given galaxy.


### Maximum likelihood - Amplitude Estimator:

$$
\begin{align}
\text{Data}: x_i & = \log{\nu_i} \\
\text{Template}: A x_0 & = A\log{\nu_o} \\
\end{align}
$$
The PDF of the magnitude is:

$$
P(x_i|A) = \frac{1}{\sqrt{2\pi}\sigma }e^{\frac{-(x_i-Ax_o)^2}{2\sigma^2}} 
$$

The Probability to get the sample of each galaxy given $A$ is:

$$
P(x_1,...,x_n|A)=\left(\sqrt{2\pi}\sigma\right) ^{-n} \ \Pi_i^n  \left (e^{\frac{-(x_i-Ax_o)^2}{2\sigma^2}} \right)
$$

Taking the logarithm of the likelikood: 

$$
\ln{P(x_1,...,x_n|A)}= \frac{-n}{2} \ln{(2 \pi)} - n \ln{(\sigma)} - \frac{1}{2\sigma^2} \sum_i^n  (x_i-Ax_o)^2
$$

Maximixing the likelihood:
$$
\begin{align}
\frac{\partial \ln{P(x_1,...,x_n|A)}} {\partial A} & = -\frac{1}{\sigma^2}\sum_i^n (x_i-Ax_o)(-x_o) &= 0 \\
                                                    &  \ \ \ \ \ \sum_i^n (x_o x_i-Ax_o^2) & = 0 \\
                                                    &  \ \ \ \ \ x_0 \sum_i^n (x_i) - n A x_0^2 &= 0
\end{align}
$$
Finally, we found that MLE for the amplitude will be given by: 
$$\boxed{ \hat{A}=\frac{\sum x_i}{n \cdot x_o} =\frac{\sum \log x_i}{n \log x_o}}$$

(ii) Assume that a galaxy at redshift $z=0$ yields the measured fluxes in our five photometric bands given in `galaxy_fluxes_z=0.txt`. Use the expression of the likelihood you have derived to determine the amplitude of the template log-fluxes that best fits the observations.

````python
observed = np.loadtxt('data/galaxy_fluxes_z=0.txt')
filters = np.loadtxt('data/bandfilters.txt')
path = "data/k_nmf_derived.newdefault.fits"
temps = fits.open(path)
tspec = temps[1].data
lam = temps[11].data
template = tspec[0]
data_temp = np.zeros(5) 
for i in range(5):
    inds = np.logical_and(lam > filters[i], lam < filters[i+1])
    data_temp[i] = template[inds].mean()
mufilters = (filters[1:] + filters[:-1]) * .5
````

````python
plt.figure(figsize=(8, 4))
plt.plot(lam, template, label = 'Spectrum', color='#AE7182', linewidth=2)
plt.plot(mufilters, observed, marker='.', markersize=9, linestyle='None', \
         color='r', label = 'Observed data')
plt.plot(mufilters, data_temp, marker='.', markersize=9, linestyle='None', \
         color='b', label = 'Template - Mean per Band')
plt.legend(frameon=True, loc = 0)
plt.xlabel(r'Wavelength in $\AA$')
plt.title('Comparison: observed galaxy and template (linear)',size = 15,pad= 15)
plt.ylabel(r'Flux')
````

In log scale, we have:

````python
plt.figure(figsize=(8, 4))
plt.plot(lam, np.log10(template), label = 'Spectrum', color='gray', linewidth=2)
plt.plot(mufilters, np.log10(observed), marker='.', markersize=9, linestyle='None', \
         color='r', label = 'Observed data')
plt.plot(mufilters, np.log10(data_temp), marker='.', markersize=9, linestyle='None', \
         color='b', label = 'Template - Mean per Band')
plt.legend(frameon=True, loc = 0)
plt.xlabel(r'Wavelength in $\AA$')
plt.title('Comparison: observed galaxy and template (magnitudes)',size = 15,pad= 15)
plt.ylabel(r'Magnitudes')
````

From the plot above it is clear that the magnitudes of the template can be scaled by a single estimated amplitude, as it was proposed in the problem.

````python
#Amplitude Estimator Function:
def amplitude_estimator(data,assume):
    i=0
    amplitude=[]
    while i<np.size(data):
        amplitude.append(np.log10(data[i])/(np.log10(assume[i])))
        i=i+1;
    return np.mean(amplitude)
````

First we decided to scale the mean values (as a form of test):

````python
#Fit of the template
amplitude_estimator= amplitude_estimator(observed, data_temp)
means_fitted_temp_log = amplitude_estimator*np.log10(data_temp)
````

````python
plt.figure(figsize=(8, 4))
plt.plot(lam, np.log10(template), label = 'Template', color='gray', linewidth=2)
plt.plot(mufilters, np.log10(observed), marker='.', markersize=9, linestyle='None', \
         color='r', label = 'Observed data')
plt.plot(mufilters, np.log10(data_temp), marker='.', markersize=9, linestyle='None', \
         color='b', label = 'Template (Means)')
plt.plot(mufilters, means_fitted_temp_log, marker='+', markersize=15, linestyle='None', \
         color='k', label = 'Fitted Template (Means)')

plt.text(15000, -13, r"$\hat{A}$ = "+str(amplitude_estimator), fontsize=15,
        ha='left', va='bottom', color="Blue")
plt.title('Magnitudes: Fitting per Bands',size = 15,pad= 15)
plt.legend(frameon=True, loc = 0)
plt.xlabel(r'Wavelength in $\AA$')
plt.ylabel(r'Log Flux - Magnitudes')
````

Here we can see that the fitting of the template's means is visually accurate, so we proceed to fit the complete spectrum.

````python
fitted_temp_log = amplitude_estimator*np.log10(template)
````

````python
plt.figure(figsize=(8, 4))
plt.plot(lam, np.log10(template), label = 'Template', color='gray', linewidth=2)
plt.plot(lam, fitted_temp_log, label = 'Fitted Template', color='k', linewidth=2)
plt.plot(mufilters, np.log10(observed), marker='.', markersize=9, linestyle='None', \
         color='r', label = 'Observed data')

plt.legend(frameon=True, loc = 0)
plt.title('Magnitudes: Fitting template spectrum to observations',size = 15,pad= 15)
plt.text(16000, -13, r"$\hat{A}$ = "+str(amplitude_estimator), fontsize=15,
         ha='left', va='bottom', color="Blue")
plt.xlabel(r'Wavelength in $\AA$')
plt.ylabel(r'Magnitudes')
````

````python
plt.figure(figsize=(8, 4))
plt.plot(lam, template, label = 'Template', color='gray', linewidth=2)
plt.plot(lam, 10**(fitted_temp_log), label = 'Fitted Template', color='k', linewidth=2)
plt.plot(mufilters, observed, marker='.', markersize=9, linestyle='None', \
         color='r', label = 'Observed data')

plt.legend(frameon=True, loc = 0)
plt.title('Fluxes: Fitting template spectrum to observations',size = 15,pad= 15)
plt.xlabel(r'Wavelength in $\AA$')
plt.ylabel(r'Fluxs')
````

We found that the value of amplitude for this galaxy is $\hat{A} = 0.97$

## Problem 2: Photometric redshifts without error model

We now extend our model from problem 1 to determine galaxy redshifts: we can estimate a galaxy's redshift by solving the maximum likelihood problem from our first exercise on a grid of redshifts. The redshift which corresponds to the solution with the smallest residuals is our maximum-likelihood redshift estimate for the galaxy.

(i) Estimate the redshift of all galaxies in catalog `catalog1.txt`, assuming that all galaxies are of type `tspec[0]` and that their redshifts lie between $0 \leq z \leq 1.5$. 


### Galaxy Redshift Estimation

This code estimates the redshifts of galaxies using a template-fitting approach. Here's a breakdown of the main components:

#### Data Loading and Setup

```python
import numpy as np
from astropy.io import fits

observed = np.loadtxt('data/galaxy_fluxes_z=0.txt')
filters = np.loadtxt('data/bandfilters.txt')
path = "data/k_nmf_derived.newdefault.fits"

lam = temps[11].data
template = tspec[0]
mufilters = (filters[1:] + filters[:-1]) * 0.5
```

This section loads the observed galaxy fluxes, filter information, and spectral templates.

#### Helper Functions

##### Amplitude Estimator

```python
def amplitude_estimator(data, assume, n):
    amplitude = np.log10(data) / (n * np.log10(assume))
    return np.mean(amplitude)
```

This function estimates the amplitude (scaling factor) between observed and model fluxes.

Equation: $A = \frac{1}{N} \sum_{i=1}^{N} \frac{\log_{10}(F_{obs,i})}{n \log_{10}(F_{model,i})}$

Where:
- $A$ is the amplitude
- $F_{obs,i}$ is the observed flux in filter i
- $F_{model,i}$ is the model flux in filter i
- $n$ is a power-law index (usually 1)
- $N$ is the number of filters

##### Chi-Square Calculation

```python
def calculate_chi_square(observed, model):
    return np.sum((observed - model)**2 / model)
```

This function calculates the chi-square statistic to measure the goodness of fit.

Equation: $\chi^2 = \sum_{i=1}^{N} \frac{(F_{obs,i} - F_{model,i})^2}{F_{model,i}}$

#### Redshift Estimation

```python
def estimate_redshift(galaxy_data, z_grid, template, lam, filters):
    min_chi_square = np.inf
    best_z = None
    
    for z in z_grid:
        # Redshift the template
        lam_z = lam * (1 + z)
        
        # Calculate the model fluxes for this redshift
        model_fluxes = np.zeros(5)
        for i in range(5):
            inds = np.logical_and(lam_z > filters[i], lam_z < filters[i+1])
            model_fluxes[i] = template[inds].mean()
        
        # Estimate amplitude
        amplitude = amplitude_estimator(galaxy_data, model_fluxes, 1)
        
        # Apply amplitude to model fluxes
        fitted_model = amplitude * model_fluxes
        
        # Calculate chi-square
        chi_square = calculate_chi_square(galaxy_data, fitted_model)
        
        if chi_square < min_chi_square:
            min_chi_square = chi_square
            best_z = z
    
    return best_z
```

This function estimates the redshift for a single galaxy by:
1. Redshifting the template: $\lambda_{z} = \lambda * (1 + z)$
2. Calculating model fluxes in each filter
3. Estimating the amplitude
4. Calculating the chi-square
5. Finding the redshift with the minimum chi-square

#### Main Execution

```python
catalog = np.loadtxt('data/catalog1.txt')
z_grid = np.linspace(0, 1.5, 1000)

estimated_redshifts = []
for galaxy in catalog:
    z_est = estimate_redshift(galaxy, z_grid, template, lam, filters)
    estimated_redshifts.append(z_est)

estimated_redshifts = np.array(estimated_redshifts)
```

This section loads the galaxy catalog, defines a redshift grid, and estimates the redshift for each galaxy in the catalog.

The code uses a maximum likelihood approach, which is equivalent to minimizing the chi-square statistic. The redshift that produces the lowest chi-square value is considered the best estimate for each galaxy.


````python
def amplitude_estimator(data, assume):
    amplitude = np.log10(data) / np.log10(assume)
    return np.mean(amplitude)
````

````python
def calculate_chi_square(observed, model):
    return np.sum((observed - model)**2 / model)
````

````python
def estimate_redshift(galaxy_data, z_grid, template, lam, filters):
    min_chi_square = np.inf
    best_z = None
    
    for z in z_grid:
        # Redshift the template
        lam_z = lam * (1 + z)
        
        # Calculate the model fluxes for this redshift
        model_fluxes = np.zeros(5)
        for i in range(5):
            inds = np.logical_and(lam_z > filters[i], lam_z < filters[i+1])
            model_fluxes[i] = template[inds].mean()
            
        # Estimate amplitude
        amplitude = amplitude_estimator(galaxy_data, model_fluxes)
        
        # Apply amplitude to model fluxes
        fitted_model = amplitude * model_fluxes
        
        # Calculate chi-square
        chi_square = calculate_chi_square(galaxy_data, fitted_model)
        
        if chi_square < min_chi_square:
            min_chi_square = chi_square
            best_z = z
    
    return best_z
````

````python
# Load galaxy catalog
catalog = np.loadtxt('data/catalog1.txt')

# Define redshift grid
z_grid = np.linspace(0, 1.5, 1000)
````

````python
# Estimate redshifts for all galaxies
estimated_redshifts = []
for galaxy in catalog:
    z_est = estimate_redshift(galaxy, z_grid, template, lam, filters)
    estimated_redshifts.append(z_est)

# Convert to numpy array
estimated_redshifts = np.array(estimated_redshifts)
````

```python
# Print statistics

print(f"Median absolute difference: {np.median(np.abs(redshift_differences))}")
print(f"Standard deviation of differences: {np.std(redshift_differences)}")
```
This section calculates and prints three key statistics:

1. Mean Absolute Difference:
   Equation: $MAD = \frac{1}{N} \sum_{i=1}^{N} |z_{est,i} - z_{true,i}|$
   
   This gives the average magnitude of the errors, regardless of whether they're overestimates or underestimates.

2. Median Absolute Difference:
   This is the middle value when all absolute differences are sorted. It's less sensitive to outliers than the mean.

3. Standard Deviation of Differences:
   Equation: $\sigma = \sqrt{\frac{1}{N} \sum_{i=1}^{N} (z_{est,i} - z_{true,i} - \mu)^2}$
   
   Where $\mu$ is the mean of the differences. This measures the spread of the errors.

These statistics help us understand:
- How accurate our estimates are on average (mean and median)
- How consistent our estimates are (standard deviation)

A lower value for each of these statistics indicates better performance:
- A low mean/median absolute difference suggests our estimates are close to the true values.
- A low standard deviation suggests our errors are consistently small across different galaxies.

By examining these numbers, we can assess how well our redshift estimation method performs and potentially identify areas for improvement.

````python
# Load true redshifts and compare
true_redshifts = np.loadtxt('data/redshifts1.txt')

# Calculate differences
redshift_differences = estimated_redshifts - true_redshifts

# Print statistics
print(f"Mean absolute difference: {np.mean(np.abs(redshift_differences))}")
print(f"Median absolute difference: {np.median(np.abs(redshift_differences))}")
print(f"Standard deviation of differences: {np.std(redshift_differences)}")
````
