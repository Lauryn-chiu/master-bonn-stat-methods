## Problem 1: Finding the best-fit template spectrum amplitude

Let us assume that in restframe, all galaxies have a spectrum as the one given in `tspec[0]` but they vary in their total luminosity. Observationally we measure galaxy magnitudes, i.e. the logarithm of observed galaxy fluxes, in the five photometric bands given above. 

In this first problem, we will assume that we do not have a model for the measurement errors on galaxy magnitudes. As a first guess, we can therefore assume that the magnitudes (i.e. the log of the flux) are Gaussian-distributed with equal errors in each photometric band. Allowing for a variable galaxy luminosity in this model amounts to introducing a constant amplitude $A$ that consistently scales the model log-fluxes in all photometric bands.

(i) Write down an analytic expression for the likelihood that needs to be minimzed to find the best-fit amplitude $A$ for a given galaxy.


### Maximum likelihood - Amplitude Estimator:

$$
\begin{align}
\text{Data}: x_i & = \log{\nu_i} \\
\text{Template}: A x_0 & = A\log{\nu_o} \\
\end{align}
$$
The PDF of the magnitude is:

$$
P(x_i|A) = \frac{1}{\sqrt{2\pi}\sigma }e^{\frac{-(x_i-Ax_o)^2}{2\sigma^2}} 
$$

The Probability to get the sample of each galaxy given $A$ is:

$$
P(x_1,...,x_n|A)=\left(\sqrt{2\pi}\sigma\right) ^{-n} \ \Pi_i^n  \left (e^{\frac{-(x_i-Ax_o)^2}{2\sigma^2}} \right)
$$

Taking the logarithm of the likelikood: 

$$
\ln{P(x_1,...,x_n|A)}= \frac{-n}{2} \ln{(2 \pi)} - n \ln{(\sigma)} - \frac{1}{2\sigma^2} \sum_i^n  (x_i-Ax_o)^2
$$

Maximixing the likelihood:
$$
\begin{align}
\frac{\partial \ln{P(x_1,...,x_n|A)}} {\partial A} & = -\frac{1}{\sigma^2}\sum_i^n (x_i-Ax_o)(-x_o) &= 0 \\
                                                    &  \ \ \ \ \ \sum_i^n (x_o x_i-Ax_o^2) & = 0 \\
                                                    &  \ \ \ \ \ x_0 \sum_i^n (x_i) - n A x_0^2 &= 0
\end{align}
$$
Finally, we found that MLE for the amplitude will be given by: 
$$\boxed{ \hat{A}=\frac{\sum x_i}{n \cdot x_o} =\frac{\sum \log x_i}{n \log x_o}}$$

(ii) Assume that a galaxy at redshift $z=0$ yields the measured fluxes in our five photometric bands given in `galaxy_fluxes_z=0.txt`. Use the expression of the likelihood you have derived to determine the amplitude of the template log-fluxes that best fits the observations. 


```python
import math
import numpy as np
from scipy.optimize import minimize
from astropy.io import fits
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
```

```python
#We load the data in the observed array
observed = np.loadtxt('data/galaxy_fluxes_z=0.txt')

filters = np.loadtxt('data/bandfilters.txt')
path = "data/k_nmf_derived.newdefault.fits"
temps = fits.open(path)
tspec = temps[1].data
lam = temps[11].data
template = tspec[0]
data_temp = np.zeros(5) 
for i in range(5):
    inds = np.logical_and(lam > filters[i], lam < filters[i+1])
    data_temp[i] = template[inds].mean()
mufilters = (filters[1:] + filters[:-1]) * .5
```

```python
plt.figure(figsize=(8, 4))
plt.plot(lam, template, label = 'Spectrum', color='#AE7182', linewidth=2)
plt.plot(mufilters, observed, marker='.', markersize=9, linestyle='None', \
         color='r', label = 'Observed data')
plt.plot(mufilters, data_temp, marker='.', markersize=9, linestyle='None', \
         color='b', label = 'Template - Mean per Band')
plt.legend(frameon=True, loc = 0)
plt.xlabel(r'Wavelength in $\AA$')
plt.title('Fluxes: Comparison between Observed galaxy and Template',size = 15,pad= 15)
plt.ylabel(r'Flux')
```

```python
#Amplitude Estimator Function:
def amplitude_estimator(data,assume,n):
    i=0
    amplitude=[]
    while i<np.size(data):
        amplitude.append(np.log10(data[i])/(n*np.log10(assume[i])))
        i=i+1;
    return np.mean(amplitude)
```

```python
#Fit of the template
amplitude_estimator= amplitude_estimator(observed, data_temp,1)
means_fitted_temp_log = amplitude_estimator*np.log10(data_temp)
fitted_temp_log = amplitude_estimator*np.log10(template)
```

```python
plt.figure(figsize=(8, 4))
plt.plot(lam, np.log10(template), label = 'Template', color='gray', linewidth=2)
plt.plot(mufilters, np.log10(observed), marker='.', markersize=9, linestyle='None', \
         color='r', label = 'Observed data')
plt.plot(mufilters, np.log10(data_temp), marker='.', markersize=9, linestyle='None', \
         color='b', label = 'Mean per Bands - Template')
plt.plot(mufilters, means_fitted_temp_log, marker='+', markersize=15, linestyle='None', \
         color='k', label = 'Mean per Bands - Template fitted')

plt.text(15000, -13, r"$\hat{A}$ = "+str(amplitude_estimator), fontsize=15,
        ha='left', va='bottom', color="Blue")
plt.title('Magnitudes: Fitting per Bands',size = 15,pad= 15)
plt.legend(frameon=True, loc = 0)
plt.xlabel(r'Wavelength in $\AA$')
plt.ylabel(r'Log Flux - Magnitudes')
```

```python
plt.figure(figsize=(8, 4))
plt.plot(lam, np.log10(template), label = 'Template', color='gray', linewidth=2)
plt.plot(lam, fitted_temp_log, label = 'Fitted Template', color='k', linewidth=2)
plt.plot(mufilters, np.log10(observed), marker='.', markersize=9, linestyle='None', \
         color='r', label = 'Observed data')

plt.legend(frameon=True, loc = 0)
plt.title('Magnitudes: Fitting template spectrum to observations',size = 15,pad= 15)
plt.text(16000, -13, r"$\hat{A}$ = "+str(amplitude_estimator), fontsize=15,
         ha='left', va='bottom', color="Blue")
plt.xlabel(r'Wavelength in $\AA$')
plt.ylabel(r'Magnitudes')
```

```python
plt.figure(figsize=(8, 4))
plt.plot(lam, template, label = 'Template', color='gray', linewidth=2)
plt.plot(lam, 10**(fitted_temp_log), label = 'Fitted Template', color='k', linewidth=2)
plt.plot(mufilters, observed, marker='.', markersize=9, linestyle='None', \
         color='r', label = 'Observed data')

plt.legend(frameon=True, loc = 0)
plt.title('Fluxes: Fitting template spectrum to observations',size = 15,pad= 15)
plt.xlabel(r'Wavelength in $\AA$')
plt.ylabel(r'Fluxs')
```
