# P04: Maximum likelihood methods

In this challenge, we are going to study the determination of photometric redshifts in galaxy surveys. 

Galaxy redshifts are usually determined by comparing the observed galaxy spectrum to templates and inferring the redshift from the observed wavelength shift of the observed spectrum. Redshifts determined in this way are called *spectroscopic redshifts*. 

Since taking galaxy spectra is time-consuming, large surveys usually cannot afford to measure the full galaxy spectrum, but only the observed flux in discrete wavelength bands. Such photometric galaxy redshifts are then inferred by comparing the observed integrated flux in these bands to that expected for different templates. 

Photometric redshifts are more noisy (and more prone to biases) than spectroscopic ones, but since they require less observing time, photometric surveys are able to observe much larger galaxy numbers. This is particularly important for surveys focused on weak gravitational lensing, as for example the recently completed [Dark Energy Survey (DES)](https://www.darkenergysurvey.org/). The photometric bands for DES cover the visible and a bit of the infrared spectrum and are shown in the figure below.

## Problem 1: Finding the best-fit template spectrum amplitude

Let us assume that in restframe, all galaxies have a spectrum as the one given in `tspec[0]` but they vary in their total luminosity. Observationally we measure galaxy magnitudes, i.e. the logarithm of observed galaxy fluxes, in the five photometric bands given above. 

In this first problem, we will assume that we do not have a model for the measurement errors on galaxy magnitudes. As a first guess, we can therefore assume that the magnitudes (i.e. the log of the flux) are Gaussian-distributed with equal errors in each photometric band. Allowing for a variable galaxy luminosity in this model amounts to introducing a constant amplitude $A$ that consistently scales the model log-fluxes in all photometric bands.

## Problem 3: Photometric redshifts with error model

Let us now assume that we have a model for our errors, i.e. we can find a full expression for our likelihood.

(i) Estimate the redshifts of the galaxies in catalogs `catalog1.txt`, `catalog2.txt`, and `catalog3.txt` assuming a Gaussian error on the log of the flux in each band of $(\sigma_g, \sigma_r, \sigma_i, \sigma_z, \sigma_Y) = (0.15, 0.1, 0.08, 0.05, 0.03)$ for catalog 1, $(\sigma_g, \sigma_r, \sigma_i, \sigma_z, \sigma_Y) = (1.0, 0.8, 0.6, 0.4, 0.2)$ for catalog 2 and $(\sigma_g, \sigma_r, \sigma_i, \sigma_z, \sigma_Y) = (0.5, 0.4, 0.3, 0.2, 0.1)$ for catalog 3. 

(ii) Analyze your results by plotting data and best-fit model with errorbars for a random selection of galaxies from the catalog. What do you conclude about the error model for catalogs `catalog1.txt`, `catalog2.txt`, and `catalog3.txt`?

(i) With error model, likelihood $$ P(x | \mu, \sigma) = \prod_{i=1}^{5} \frac{1}{\sqrt{2\pi}\sigma_i} e^{-\frac{(x_i - \mu_i)^2}{2\sigma_i^2}}$$
$\sigma_i$ is different in different photometric bands.


To maximize likelihood, we can maximize log P, where
$$ log P = \sum_{i=1}^{5} - \log(\sqrt{2\pi}\sigma_i) - \frac{(x_i - \mu_i)^2}{2\sigma_i^2},$$
Take derivative
$$ \frac{\partial \log P}{\partial \mu} = 0 $$
$$ \sum_{i=1}^{5} (-1)\frac{(-1) \cdot 2 \cdot (x_i - \mu_i)}{2\sigma_i^2} = \sum_{i=1}^{5} \frac{x_i - \mu_i}{\sigma_i^2} = 0 $$
where $x_i$ is the mean flux magnitude in an observed sample galaxy in catalogs, $\mu_i = A x_{m,i}$, $x_{m,i}$ denotes the mean flux magnitude in i-th band in template spectrum ($tspec[0]$)
$$ \sum_{i=1}^{5} \frac{x_i - A \cdot x_{m,i}}{\sigma_i^2} = 0 \Rightarrow A = \frac{\sum_{i=1}^{5} (x_i/\sigma_i^2)}{\sum_{i=1}^{5} (x_{m,i}/\sigma_i^2)} $$


```python
# import galaxy catalogs, take log to get magnitudes
catalog1 = np.loadtxt('data/catalog1.txt')
gal_mag_catalog1 = np.log10(catalog1)
catalog2 = np.loadtxt('data/catalog2.txt')
gal_mag_catalog2 = np.log10(catalog2)
catalog3 = np.loadtxt('data/catalog3.txt')
gal_mag_catalog3 = np.log10(catalog3)
# import true redshifts
true_redshift1 = np.loadtxt('data/redshifts1.txt')
true_redshift2 = np.loadtxt('data/redshifts2.txt')
true_redshift3 = np.loadtxt('data/redshifts3.txt')
```

```python
# compute scale A and then model points
def z_model_with_err(z, lam, template, gal_mag, gauss_err):
    redshift_filters = filters / (1+z) 
    data = np.zeros(5)
    for i in range(5):
        inds = np.logical_and(lam > redshift_filters[i], lam < redshift_filters[i+1])
        data[i] = template[inds].mean()
        mag_data = np.log10(data)
        scale_a_with_err = np.sum(gal_mag/gauss_err**2) / np.sum(mag_data/gauss_err**2)
        data_estimate_with_err = scale_a_with_err * mag_data
        
    return data_estimate_with_err
    
# plug in a grid of z values and return the z with minimum residual value (|value in catalog - value predicted from model|)
def optm_z_with_err(z, lam, template, gal_mag, gauss_err):
    z_optim = np.zeros(z.size)
    for i in np.arange(z.size):
        z_optim[i] = np.sum(np.abs(z_model_with_err(z[i], lam, template, gal_mag, gauss_err) - gal_mag))
        
    return np.argmin(z_optim)
```

```python
# define Gaussian errors for 5 bands in catalog 1, 2, and 3
error_array=np.array([[0.15,0.1,0.08,0.05,0.03],[1.0,0.8,0.6,0.4,0.2],[0.5,0.4,0.3,0.2,0.1]])
```

```python
# calculate redshift estimation for catalog1
cat1_redshift_est_with_err = np.zeros(gal_mag_catalog1.shape[0])
for i in np.arange(gal_mag_catalog1.shape[0]):
    gal_mag = gal_mag_catalog1[i]
    z_guess=np.linspace(0, 1.5, 300)
    cat1_redshift_est_with_err[i] = z_guess[optm_z_with_err(z_guess, lam, template, gal_mag, error_array[0,:])]
    
```

```python
# calculate redshift estimation for catalog2
cat2_redshift_est_with_err = np.zeros(gal_mag_catalog2.shape[0])
for i in np.arange(gal_mag_catalog2.shape[0]):
    gal_mag = gal_mag_catalog2[i]
    z_guess=np.linspace(0, 1.5, 300)
    cat2_redshift_est_with_err[i] = z_guess[optm_z_with_err(z_guess, lam, template, gal_mag, error_array[1,:])]
    
```

```python
# calculate redshift estimation for catalog3
cat3_redshift_est_with_err = np.zeros(gal_mag_catalog3.shape[0])
for i in np.arange(gal_mag_catalog3.shape[0]):
    gal_mag = gal_mag_catalog3[i]
    z_guess=np.linspace(0, 1.5, 300)
    cat3_redshift_est_with_err[i] = z_guess[optm_z_with_err(z_guess, lam, template, gal_mag, error_array[2,:])]
    
```

```python
# calculate the difference between redshift estimation and true redshift
redshift_deviation1 = true_redshift1 - cat1_redshift_est_with_err
redshift_deviation2 = true_redshift2 - cat2_redshift_est_with_err
redshift_deviation3 = true_redshift3 - cat3_redshift_est_with_err

plt.figure(figsize=(12,4))

plt.subplot(1, 3, 1)
plt.hist(redshift_deviation1, bins=30)
plt.axvline(x=0.0, color='k')
plt.title('z deviation in catalog1')
plt.xlabel('z difference')
plt.ylabel('count')
plt.xlim(-1,1)

plt.subplot(1, 3, 2)
plt.hist(redshift_deviation2, bins=30)
plt.axvline(x=0.0, color='k')
plt.title('z deviation in catalog2')
plt.xlabel('z difference')
plt.ylabel('count')
plt.xlim(-1,1)

plt.subplot(1, 3, 3)
plt.hist(redshift_deviation3, bins=30)
plt.axvline(x=0.0, color='k')
plt.title('z deviation in catalog3')
plt.xlabel('z difference')
plt.ylabel('count')
plt.xlim(-1,1)

plt.show()
```

```python
gal_ind1 = np.random.randint(0, 1000) # pick random galaxy from catalog1
gal_ind2 = np.random.randint(0, 1000) # pick random galaxy from catalog2
gal_ind3 = np.random.randint(0, 1000) # pick random galaxy from catalog3

plt.figure(figsize=(15,4))
plt.subplot(1, 3, 1)
plt.errorbar(mufilters, gal_mag_catalog1[gal_ind1, :], yerr=error_array[0,:], marker='.', markersize=9, linestyle='None', \
         color='#E50000', label = 'redshift galaxy', alpha=0.8)
plt.plot(mufilters, z_model_with_err(cat1_redshift_est_with_err[gal_ind1], lam, template, gal_mag_catalog1[gal_ind1], error_array[0,:]), marker='+', markersize=9, linestyle='None', \
         color='#0000FF', label = 'scaled model', alpha=0.8)
plt.axvspan(filters[0], filters[1], alpha = .2, color = 'k', zorder=-1)
plt.axvspan(filters[1], filters[2], alpha = .1, color = 'k', zorder=-1)
plt.axvspan(filters[2], filters[3], alpha = .2, color = 'k', zorder=-1)
plt.axvspan(filters[3], filters[4], alpha = .1, color = 'k', zorder=-1)
plt.axvspan(filters[4], filters[5], alpha = .2, color = 'k', zorder=-1)
plt.legend(frameon=False, loc = 0)
plt.xlabel(r'Wavelength in $\AA$')
plt.ylabel(r'Magnitude')
plt.title('random galaxy in catalog1')

plt.subplot(1, 3, 2)
plt.errorbar(mufilters, gal_mag_catalog2[gal_ind2, :], yerr=error_array[1,:], marker='.', markersize=9, linestyle='None', \
         color='#E50000', label = 'redshift galaxy')
plt.plot(mufilters, z_model_with_err(cat2_redshift_est_with_err[gal_ind2], lam, template, gal_mag_catalog2[gal_ind2], error_array[1,:]), marker='+', markersize=9, linestyle='None', \
         color='#0000FF', label = 'scaled model', alpha=0.8)
plt.axvspan(filters[0], filters[1], alpha = .2, color = 'k', zorder=-1)
plt.axvspan(filters[1], filters[2], alpha = .1, color = 'k', zorder=-1)
plt.axvspan(filters[2], filters[3], alpha = .2, color = 'k', zorder=-1)
plt.axvspan(filters[3], filters[4], alpha = .1, color = 'k', zorder=-1)
plt.axvspan(filters[4], filters[5], alpha = .2, color = 'k', zorder=-1)
plt.legend(frameon=False, loc = 0)
plt.xlabel(r'Wavelength in $\AA$')
plt.title('random galaxy in catalog2')

plt.subplot(1, 3, 3)
plt.errorbar(mufilters, gal_mag_catalog3[gal_ind3, :], yerr=error_array[2,:], marker='.', markersize=9, linestyle='None', \
         color='#E50000', label = 'redshift galaxy')
plt.plot(mufilters, z_model_with_err(cat3_redshift_est_with_err[gal_ind3], lam, template, gal_mag_catalog3[gal_ind3], error_array[2,:]), marker='+', markersize=9, linestyle='None', \
         color='#0000FF', label = 'scaled model', alpha=0.8)

plt.axvspan(filters[0], filters[1], alpha = .2, color = 'k', zorder=-1)
plt.axvspan(filters[1], filters[2], alpha = .1, color = 'k', zorder=-1)
plt.axvspan(filters[2], filters[3], alpha = .2, color = 'k', zorder=-1)
plt.axvspan(filters[3], filters[4], alpha = .1, color = 'k', zorder=-1)
plt.axvspan(filters[4], filters[5], alpha = .2, color = 'k', zorder=-1)
plt.legend(frameon=False, loc = 0)
plt.xlabel(r'Wavelength in $\AA$')
plt.title('random galaxy in catalog3')

plt.show()
```

From the histograms which plot the difference between estimated redshift from maximum likelihood and true redshift given in data folder, we can see a Gaussian-like profile, which may imply that our estimate generally predict redshifts that is close to the true values with errors behaving like a Gaussian.

From the plots of random galaxies, we see clear deviations of the scaled model prediction (blue + sign) from the observed galaxies (red dot sign), this may indicate that large error bars bring more measurement and modeling uncertainties.
